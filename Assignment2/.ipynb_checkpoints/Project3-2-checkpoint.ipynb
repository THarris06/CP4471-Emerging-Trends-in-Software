```python
#1
import pandas as pd
data = pd.read_csv('avocado.csv')
data.head(5)
```

```python
#2
df_data = data.groupby('type')[['Small Bags', 'Large Bags', 'XLarge Bags']].sum()
df_columns = ['Small Bags', 'Large Bags', 'XLarge Bags']
df = pd.DataFrame(data=df_data, columns=df_columns)
df = df.reset_index();
df
```

```python
#3
df.plot.bar()
```
